```python
import torch
import torchvision
import torch.nn as nn
import torch.nn.functional as F
from torchsummary import summary

class AlexNet(nn.Module):

    def __init__(self, n_classes):
        super(AlexNet, self).__init__()



        ############## Add Layer ##############
        self.conv1 = nn.Conv2d(3, 96, kernel_size=11, stride=4, padding=2)
        self.conv2 = nn.Conv2d(96, 256, kernel_size=5, padding=2)
        self.conv3 = nn.Conv2d(256, 384, kernel_size=3, padding=1)
        self.conv4 = nn.Conv2d(384, 384, kernel_size=3, padding=1)
        self.conv5 = nn.Conv2d(384, 256, kernel_size=3, padding=1)

        self.fc1 = nn.Linear(256 * 6 * 6, 4096)
        self.fc2 = nn.Linear(4096 , 4096)
        self.fc3 = nn.Linear(4096 , n_classes)


        #######################################

    def forward(self, x):

        ############## Add Layer ##############

        x = F.max_pool2d(F.relu(self.conv1(x)), kernel_size=3, stride=2)
        x = F.max_pool2d(F.relu(self.conv2(x)), kernel_size=3, stride=2)
        x = F.relu(self.conv3(x))
        x = F.relu(self.conv4(x))
        x = F.max_pool2d(F.relu(self.conv5(x)), kernel_size=3, stride=2)

        x = x.view(x.size(0), -1)
        x = F.relu(self.fc1(x))
        x = F.dropout(x)
        x = F.relu(self.fc2(x))
        x = F.dropout(x)
        x = self.fc3(x)




        #######################################

        return x


# pytorch summary
# summary code 추가
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
model = AlexNet(100).to(device)
summary(model, (3, 227, 227))
```
